1. Difference between interpreted and compiled languages

Interpreted languages (ex-Python): Code is executed line by line at runtime.

Compiled languages (ex-c,c++): Code is translated into machine code before execution.

2. What is exception handling in Python

Exception handling is a mechanism to handle runtime errors using try, except, else, and finally blocks.

3. Purpose of the finally block

The finally block executes always, whether an exception occurs or not, and is used for cleanup tasks.

4. What is logging in Python

Logging is used to record program events, errors, and execution details for debugging and monitoring.

5. Significance of the __del__ method

__del__ is a destructor method called when an object is about to be destroyed.

6. Difference between import and from ... import

import module: Access using module.name

from module import name: Access directly using name

7. Handling multiple exceptions in Python

Use multiple except blocks or a tuple:

8. Purpose of the with statement in file handling

It ensures files are automatically closed, even if an error occurs.

9. Difference between multithreading and multiprocessing

Multithreading: Multiple threads in the same process (shared memory)

Multiprocessing: Multiple processes with separate memory spaces

10. Advantages of using logging

Helps debug errors

Tracks program flow

Stores error history

Better than print statements

11. What is memory management in Python

Python automatically manages memory allocation and deallocation.

17. Common logging levels in Python

DEBUG
Detailed information for diagnosing problems

Used mainly during development

INFO
Confirms that the program is working as expected

Used for general informational messages

WARNING
Indicates a potential problem or unexpected situation

The program continues to run

ERROR
Indicates a serious problem that prevents part of the program from working

Used when an exception occurs

CRITICAL
Indicates a very serious error

The program may not be able to continue running
